#### Understanding IRIS dataset

```python
import seaborn as sns
```

```python
from sklearn import datasets
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn import metrics
```

```python
#Load dataset
iris = datasets.load_iris()
print(iris.target_names)
print(iris.feature_names)
```

```python
data=pd.DataFrame({'sepal length':iris.data[:,0],'sepal width':iris.data[:,1],'petal length':iris.data[:,2],'petal width':iris.data[:,3],
'species':iris.target
})
#data: 0 to 49 = setrosa 50:99 = versicolor and 100:149 = virginica
data_train1=data[0:40]
data_train2=data[50:90]
data_train3=data[100:140]
data_train=pd.concat([data_train1,data_train2,data_train3],axis=0)

data_test1=data[40:50]
data_test2=data[90:100]
data_test3=data[140:150]
data_test=pd.concat([data_test1,data_test2,data_test3],axis=0)
```

```python
sns.relplot(x="sepal length", y="sepal width", hue="species",data=data,palette="Set1")
```

```python
sns.relplot(x="petal length", y="petal width", hue="species",data=data,palette="Set1")
```

The graph shows that the 3 classes can be eaisly sepearated using patel length and petal width.Also using sklearn's *feature_importances_* method we can show that more 85% of the importance is given to these features. 

```python
X=data_train[['sepal length', 'sepal width', 'petal length', 'petal width']]  # Features
y=data_train['species']  # Labels
```

```python
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2)
```

```python
model = GaussianNB()
```

```python
model.fit(X_train,y_train)
y_pred=model.predict(X_val)
```

```python
print("Accuracy using criteria is:",metrics.accuracy_score(y_val, y_pred))# accuracy 1 indicates that the training is perfect, but we need to check if model is overfitting using test data
```

```python
X_test=data_test[['sepal length', 'sepal width', 'petal length', 'petal width']]
```

```python
print(model.predict(X_test))
```
